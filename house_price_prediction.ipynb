```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Setup Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Inisialisasi WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# Buka website
url = "https://www.rumah123.com/jual/surabaya/rumah/"
driver.get(url)
time.sleep(5)  # Tunggu halaman termuat

# Scroll untuk memastikan semua listing muncul
for _ in range(5):
    driver.execute_script("window.scrollBy(0, 500);")
    time.sleep(2)

# Ambil semua elemen listing rumah
listings = driver.find_elements(By.CLASS_NAME, "rui__sc-1mi1xlz-0")  # Sesuaikan class name

# Simpan data ke dalam list
data = []
for listing in listings:
    try:
        # Ambil informasi utama
        title = listing.find_element(By.CLASS_NAME, "rui__sc-1wdrlhx-3").text.strip()
        price = listing.find_element(By.CLASS_NAME, "rui__sc-1wdrlhx-4").text.strip()
        location = listing.find_element(By.CLASS_NAME, "rui__sc-1wdrlhx-7").text.strip()
        details = listing.find_elements(By.CLASS_NAME, "rui__sc-1wdrlhx-9")  # Semua detail rumah

        # Ambil data luas tanah dan luas bangunan
        lt = "N/A"
        lb = "N/A"
        for detail in details:
            text = detail.text
            if "LT" in text:
                lt = text.split(":")[1].strip()
            if "LB" in text:
                lb = text.split(":")[1].strip()

        # Ambil URL rumah
        link = listing.find_element(By.TAG_NAME, "a").get_attribute("href")

        # Simpan ke list
        data.append([title, price, location, lt, lb, link])

    except Exception as e:
        print(f"Error mengambil data: {e}")
        continue

# Simpan ke DataFrame
df = pd.DataFrame(data, columns=["Title", "Price", "Location", "Luas Tanah", "Luas Bangunan", "URL"])

# Simpan hasil ke CSV
df.to_csv("rumah123_surabaya1.csv", index=False)

# Tampilkan hasil scraping
print(df.head())

# Tutup browser
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL Rumah123
url = "https://www.rumah123.com/jual/surabaya/rumah/"

# Simulasikan browser agar tidak terblokir
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# Kirim permintaan GET
response = requests.get(url, headers=headers)

# Periksa apakah request berhasil
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")

    # Cari elemen yang berisi daftar rumah
    listings = soup.find_all("div", class_="rui__sc-1mi1xlz-0")

    data = []
    for listing in listings:
        try:
            # Ambil informasi dasar
            title = listing.find("h2").text.strip() if listing.find("h2") else "N/A"
            price = listing.find("div", class_="rui__sc-1wdrlhx-4").text.strip() if listing.find("div", class_="rui__sc-1wdrlhx-4") else "N/A"
            location = listing.find("div", class_="rui__sc-1wdrlhx-7").text.strip() if listing.find("div", class_="rui__sc-1wdrlhx-7") else "N/A"

            # Ambil URL listing
            link_tag = listing.find("a", href=True)
            link = "https://www.rumah123.com" + link_tag["href"] if link_tag else "N/A"

            # Simpan ke list
            data.append([title, price, location, link])

        except Exception as e:
            print(f"Error mengambil data: {e}")
            continue

    # Simpan ke DataFrame
    df = pd.DataFrame(data, columns=["Title", "Price", "Location", "URL"])

    # Simpan ke CSV
    df.to_csv("rumah123_surabaya.csv", index=False)
    print(df.head())

else:
    print(f"Request gagal dengan status code: {response.status_code}")
```

```python
import requests
import pandas as pd

# URL yang ditemukan di Network
url = "https://harga-emas.org/data.php?v_type=antam&v_range=1825"

# Mengirim permintaan GET
response = requests.get(url)

# Mengecek apakah permintaan berhasil
if response.status_code == 200:
    try:
        data = response.json()  # Konversi ke JSON jika formatnya JSON
        df = pd.DataFrame(data)  # Buat DataFrame dari data
        print("Dataframe created from JSON data")
        print(df.head())  # Tampilkan 5 data pertama
    except requests.exceptions.JSONDecodeError:
        print("Response is not in JSON format. Attempting to parse as HTML table.")
        # If JSON decoding fails, assume it's an HTML table
        df = pd.read_html(response.text)[0] #Parse the HTML response as table
        print("Dataframe created from HTML table")
        print(df.head())
else:
    print("Gagal mengambil data:", response.status_code)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL yang ditemukan di Network
url = "https://harga-emas.org/data.php?v_type=antam&v_range=1825"

# Mengirim permintaan GET
response = requests.get(url)

# Mengecek apakah permintaan berhasil
if response.status_code == 200:
    try:
        # Check if the response content type is JSON
        if 'application/json' in response.headers.get('Content-Type', ''):
            data = response.json()  # Konversi ke JSON jika formatnya JSON
            df = pd.DataFrame(data)  # Buat DataFrame dari data
            print("Dataframe created from JSON data")
            print(df.head())  # Tampilkan 5 data pertama
        else:
            print("Response is not in JSON format. Attempting to parse as HTML.")
            soup = BeautifulSoup(response.text, 'html.parser')
            # Check if there's a table to parse
            tables = soup.find_all('table')
            if tables:
                df = pd.read_html(str(tables))[0] # Parse the HTML table
                print("Dataframe created from HTML table")
                print(df.head())
            else:
                print("No tables found in the HTML response")
                print("HTML Content:",response.text) #Print the html to inspect
    except requests.exceptions.JSONDecodeError:
        print("Response is not in JSON format and could not be parsed as HTML table.")
    except Exception as e:
        print(f"An error occurred: {e}")
else:
    print("Gagal mengambil data:", response.status_code)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL sumber data
url = "https://harga-emas.org/data.php?v_type=antam&v_range=1825"

# Mengambil data dari URL
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, "xml")  # Parsing sebagai XML

    # Ambil kategori (tanggal)
    dates = [tag["name"] for tag in soup.find_all("category")]

    # Ambil harga emas (nilai dari <set>)
    prices = [tag["value"] for tag in soup.find_all("set")]

    # Pastikan panjang kedua data sama
    min_length = min(len(dates), len(prices))  # Ambil panjang minimum
    dates = dates[:min_length]  # Potong jika lebih panjang
    prices = prices[:min_length]  # Potong jika lebih panjang

    # Simpan dalam DataFrame
    df = pd.DataFrame({"Tanggal": dates, "Harga Emas (IDR)": prices})

    # Simpan ke file CSV
    csv_filename = "harga_emas.csv"
    df.to_csv(csv_filename, index=False)

    print(f"✅ Data berhasil disimpan ke '{csv_filename}'!")
else:
    print(f"⚠️ Gagal mengambil data! Status code: {response.status_code}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Membaca data dari CSV
csv_filename = "harga_emas.csv"
df = pd.read_csv(csv_filename)

# Konversi kolom "Tanggal" menjadi format datetime
df["Tanggal"] = pd.to_datetime(df["Tanggal"], format="%d %b%y")

# Plot grafik
plt.figure(figsize=(12, 6))
plt.plot(df["Tanggal"], df["Harga Emas (IDR)"], marker="o", linestyle="-", color="b", label="Harga Emas")

# Format tampilan
plt.xlabel("Tanggal")
plt.ylabel("Harga Emas (IDR)")
plt.title("Perubahan Harga Emas dari Waktu ke Waktu")
plt.xticks(rotation=45)  # Rotasi label tanggal agar lebih mudah dibaca
plt.legend()
plt.grid(True)

# Menampilkan grafik
plt.show()
```
