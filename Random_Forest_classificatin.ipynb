#Random Forest Classification


Random Forest adalah algoritma pembelajaran mesin berbasis ensemble yang digunakan untuk tugas klasifikasi dan regresi. Algoritma ini bekerja dengan membuat kumpulan decision tree selama pelatihan dan menggabungkan hasilnya untuk menghasilkan prediksi yang lebih akurat dan stabil.


#Ensemble Learning
adalah teknik dalam pembelajaran mesin yang menggabungkan beberapa model untuk meningkatkan performa prediksi dibandingkan dengan model tunggal. Ensemble learning dibagi menjadi dua kategori utama berdasarkan jenis model yang digunakan:

>1. Homogeneous Ensemble Learning
Homogeneous ensemble learning menggunakan beberapa model yang sama untuk melakukan prediksi. Semua model dalam ensemble ini dilatih dengan algoritma yang sama, namun dengan data yang berbeda atau parameter yang sedikit dimodifikasi untuk meningkatkan variabilitas.

Contoh:
Random Forest: Merupakan contoh dari homogeneous ensemble yang menggunakan banyak decision trees (pohon keputusan) yang dilatih dengan data yang berbeda dan hanya mempertimbangkan subset fitur yang berbeda pada setiap split.
Bagging (Bootstrap Aggregating): Menggunakan model yang sama (misalnya, pohon keputusan) yang dilatih pada subset data yang diambil dengan teknik bootstrap sampling. Setiap model menghasilkan prediksi, dan hasil akhirnya adalah rata-rata (regresi) atau voting mayoritas (klasifikasi).

>2. Heterogeneous Ensemble Learning
Heterogeneous ensemble learning menggunakan model yang berbeda untuk membuat prediksi. Berbeda dengan homogeneous ensemble, model-model yang digunakan di sini memiliki algoritma yang berbeda, yang dapat menangani jenis data yang berbeda atau memanfaatkan keunggulan masing-masing model.

Contoh:
AdaBoost: Menggunakan pohon keputusan yang sangat sederhana (sering disebut stumps), namun menggabungkan model-model ini dengan memberikan bobot yang lebih tinggi pada kesalahan yang dibuat sebelumnya untuk meningkatkan performa prediksi.
Voting Classifier: Menggunakan beberapa model yang berbeda, misalnya logistic regression, random forest, dan SVM, kemudian melakukan voting mayoritas atau rata-rata untuk memutuskan hasil akhir.
Stacking: Menggabungkan beberapa model yang berbeda dan melatih model final (meta-model) untuk memprediksi hasil akhir berdasarkan output dari model-model sebelumnya.

#Bagging (Bootstrap Aggregating)
 adalah teknik ensemble learning yang digunakan untuk meningkatkan stabilitas dan akurasi dari model pembelajaran mesin, dengan cara menggabungkan beberapa model yang dilatih pada subset data yang berbeda. Teknik ini sangat efektif untuk mengurangi variansi dan menghindari overfitting, khususnya pada model yang cenderung overfit, seperti decision trees.

#dataset

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print("Dimensi Feature:", X.shape)
print("Class:", set(y))
```

#split dataset

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=0
                                                    )
```

#klasifikasi dengan random forest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100,
                               random_state=0)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
